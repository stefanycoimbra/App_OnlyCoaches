## Importing some libraries
We need to import the Urllib and BeautifulSoup to do the data collection and parse the URL content of data, respectively. Then, we need to import NLTK library to do the tokenization and calculate the word frequency for each word. Here, we are also importing punctuation from string to remove the puncuation from the text to do the summarization. The 're' is the library that we will be using to treat the text and substitute some unnecessary parts of the article.

```python
import urllib
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize 
from string import punctuation
import re

nltk.download('punkt')
nltk.download("stopwords")
stop_words = stopwords.words('english')
```



---

##Saving the URL from the webpage and reading the HTML of it

```python
url = "https://en.wikipedia.org/wiki/Artificial_intelligence"
```

```python
webpage = urllib.request.urlopen(url)
```

```python
article = webpage.read()
```

```python
article
```



---

##Using BeautifulSoup to parse the HTML content of the data

```python
article_parsed = BeautifulSoup(article,'html.parser')
```

We need only the main text, right? That's why we need to find all 'p' tags in the HTML and we will extract the text from them:

```python
paragraphs = article_parsed.find_all('p')
```

We go through all the paragraphs and to each one, we concatenate It to a string named article_content:

```python
article_content = ''
for p in paragraphs:  
    article_content += p.text
```

Let's see the format of our article until now below. There are a lot of brackets with numbers that refers to the bibliography in the webpage, but we don't want that and we want our text as clean as possible so we can do the summary without problems. That's why we use re.sub(), to substitute every bracket's content to only brackets. Then, using translate() function with str.maketrans() we can remove the brackets for good and the tag '\xa0' that appers during the reading and It isn't convenient:

```python
article_content
```

```python
article_replaced = re.sub("\[.*?\]","[]", article_content)
```

```python
article_replaced = article_replaced.translate(str.maketrans({'[': '', ']': '', '\xa0':''}))
```

Now our article is cleaner and we can start working in our summarization!

```python
article_replaced
```



---

##Tokenization time! We will tokenize the text to extract word by word

```python
tokens = word_tokenize(article_replaced)
```



---

##Removing punctuation using the module from String's library

Here we add the '\n' so we can put all the text together:

```python
punctuation = punctuation + '\n'
punctuation
```



---

##Calculating the words' frequencies

```python
words_frequencies = {}
for word in tokens:    
    if word.lower() not in stop_words and word.lower() not in punctuation:
       # if the word is not in the dictionary yet
       if word not in words_frequencies.keys():
           words_frequencies[word] = 1
       # otherwise, we increase the counting for the word
       else:
           words_frequencies[word] += 1
```

We notice that the word that most appears in the text is 'AI', with 125 occurrences:

```python
words_frequencies
```

As we saw, the higher frequency is 125, that is 'AI' word:

```python
max_frequency = max(words_frequencies.values())
print(max_frequency)
```



---

##Once we tokenized to extract word by word, now we will separate sentence by sentence

```python
sent_token = sent_tokenize(article_replaced)
```

```python
sent_token
```



---

##Calculating the sentences' scores

```python
sentences_scores = {}
for sent in sent_token:
    # we split the sentences
    sentence = sent.split(" ")
    for word in sentence:        
        if word.lower() in words_frequencies.keys():
            # if it's a new sentence, we set the score of the first word
            if sent not in sentences_scores.keys():
                sentences_scores[sent] = words_frequencies[word.lower()]
            # otherwise we sum the score of the word with the scores added before
            else:
                sentences_scores[sent] += words_frequencies[word.lower()]
```

We can see the total score to each sentence now:

```python
sentences_scores
```



---

##Calculating the weighted frequency for each sentence based on the maximum frequency

```python
for word in words_frequencies.keys():
    words_frequencies[word] = (words_frequencies[word]/max_frequency)
```

```python
words_frequencies
```



---

##Summarization

First, let's see how many sentences are if we want only 30% of the text. To do that we simply calculate 0.3*len(sent_token), because sent_token contains all sentences:

```python
total_of_sentences = len(sent_token)*0.3
total_of_sentences
```

It's not an integer number, so let's cast It to an integer:

```python
total_of_sentences = int(total_of_sentences)
total_of_sentences
```

Now, to build the summary, we will use the method nlargest from heapq, because with It we can use the sentences' scores and total of sentences that we already calculated:

```python
from heapq import nlargest
```

```python
summary = nlargest(total_of_sentences, sentences_scores, key = sentences_scores.get)
```

Let's put all the sentences from the summary together to finally get our final answer:

```python
final_summary = [word for word in summary]
summary = ' '.join(final_summary)
```

```python
summary
```
